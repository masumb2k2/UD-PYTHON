```python
# Method Overriding
# Base class
class Animal:
    def speak(self):
        return "sound the animal"
    
# derive class 
class Dog(Animal):
    def speak(self):
        return "woof!"
# derived class
class Cat(Animal):
    def speak(self):
        return "Meaw"
# function that demonstrate polymorphism
def animal_speak(animal):
    print(animal.speak())

dog=Dog()
cat=Cat()
print(dog.speak())
print(cat.speak())
animal_speak(dog)
```

```python
# Polumorhism with function and methods
# base class 
class Shape:
    def area(self):
        return "the are of the figure"
# Derived class 1
class Rectangle(Shape):
    def __init__(self,Width,height):
        self.width=Width
        self.height=height
    def area(self):
        return self.height*self.width
# derived class 2 
class Circle(Shape):
    def __init__(self,radious):
        self.radious=radious
    def area(self):
        return self.radious*3.14*self.radious

# function that demostrate ploy
def print_area(shape):
    print(f"the area is {shape.area()}")

# object
rectangle=Rectangle(12,5)
circle=Circle(12)

print_area(rectangle)
print_area(circle)
```

```python
# Abstract base classes
from abc import ABC,abstractmethod
# define abstract class
class vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass
# Derived class 1
class Car(vehicle):
    def start_engine(self):
        return "car engine started"
# derived class 2
class Cycle(vehicle):
    def start_engine(self):
        return "Cycle Started"
def start_vehicle(vcle):
    print(vcle.start_engine())
#create object 
car1=Car()
cycle1=Cycle()
start_vehicle(car1)
```
