```python
set1={'IronMan',"Avengers",'Hitman'}
set2={'IronMan',"Avengers",'Hitman','Hulk2'}
```

```python
# Difference 
set2.difference(set1)
```

```python
set2
```

```python
# Difference Update 
set2.difference_update(set1)
```

```python
set2
```

```python
my_dic={'car1':'Audi','car2':'BMW','car3':'Benz'}
```

```python
my_dic['car1']
```

```python
for x in my_dic:# using key
    print(x)
```
