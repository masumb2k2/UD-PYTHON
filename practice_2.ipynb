# Second Practice of python  

```python
# Conditional statement in python
mark=int(input('ENter your mark: '))

if mark>0 and mark<40:
    print('You are Fail')
elif mark>=40 and mark<80:
    print('Pass but not A+')
elif mark>=80 and mark<=100:
    print("Pass with A+")
else:
    print('Invalid Input')
```

# For Loop in Python

```python
for i in range(10):
    print(i)
```

```python
for i in range (10,19,3):
    print(f'Value is: {i}')
```

```python
name='Masum Billah'
for char in name:
    print(f'Character is {char}')
```

# while Loop

```python
i=0
while i<=10:
    print('Value is ', i)
    i+=1
```

```python
# break statement
i=1
while i<=10:
    print('Value is', i)
    if i==7:
        break
    i+=1
```

```python
# Continue Statement 
i=1
while i<=20:
    i+=1
    if i%2==0:
        continue
    print(i)
    
```

```python
# pass statement
i =0
while i<=10:
    i+=1
    print(i)
    if i==3:
        print('inside pass')
        pass
```
