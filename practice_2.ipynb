# while Loop

```python
i=0
while i<=10:
    print('Value is ', i)
    i+=1
```

```python
# break statement
i=1
while i<=10:
    print('Value is', i)
    if i==7:
        break
    i+=1
```

```python
# Continue Statement 
i=1
while i<=20:
    i+=1
    if i%2==0:
        continue
    print(i)
    
```

```python
# pass statement
i =0
while i<=10:
    i+=1
    print(i)
    if i==3:
        print('inside pass')
        pass
```

```python
for key in dic.keys():
    print('Key is :',key)
```
