# Nested Dictionary

```python
car1_model={'Mercedes':1960}
car2_model={'Audi': 1970}
car3_model={'Ambassador':1980}

car_type={'car1':car1_model,'car2':car2_model,'car3':car3_model}
```

```python
print(car_type)
```

```python
print(car_type['car1'])
```

```python
print(car_type['car1']['Mercedes'])
```
