# Condtitional Operator

```python
temp=int(input("Enter Temperature in Celcious:"))
if temp>0 and temp<=25:
    print(f'the weather is cold and the temperature is {temp} degree celcious')
elif temp>25 and temp <=35:
    print(f'the weather is moderate and the temperature is {temp} degree celcious')
elif temp>35:
    print(f'the weather is hot and the temperature is {temp} degree celcious')
else:
    print('Invalid Input')
```

# LOOPs

```python
# for loop 
limit=int(input("Enter the limit:"))
for i in range(limit):
    print(f'The current value is {i}')
```

```python
limit=int(input("Enter the limit:"))
for i in range(0,limit,2):
    print(f'The current value is {i}')
```

```python
# While Loop
limit=int(input("Enter the Value of Limit:"))
i=0
while i<=limit:
    print('Current Value is: ',i)
    i+=1
```

```python
i=0
while i<=10:
    print('Current Value is: ',i)
    if i==7:
        break
    i+=1
```

```python
i=0
while i<=20:
    i+=1
    if i%2==0:
        continue
    print ('Current value is ',i)
    
```

```python
# Use of pass 
for i in range(1,20,1):
    if i%2==0:
        print('inside pass value ',i)
        pass
    print(i)
```
