# Condtitional Operator

```python
temp=int(input("Enter Temperature in Celcious:"))
if temp>0 and temp<=25:
    print(f'the weather is cold and the temperature is {temp} degree celcious')
elif temp>25 and temp <=35:
    print(f'the weather is moderate and the temperature is {temp} degree celcious')
elif temp>35:
    print(f'the weather is hot and the temperature is {temp} degree celcious')
else:
    print('Invalid Input')
```

# LOOPs

```python
# for loop 
limit=int(input("Enter the limit:"))
for i in range(limit):
    print(f'The current value is {i}')
```

```python
limit=int(input("Enter the limit:"))
for i in range(0,limit,2):
    print(f'The current value is {i}')
```

```python
# While Loop
limit=int(input("Enter the Value of Limit:"))
i=0
while i<=limit:
    print('Current Value is: ',i)
    i+=1
```

```python
i=0
while i<=10:
    print('Current Value is: ',i)
    if i==7:
        break
    i+=1
```

```python
i=0
while i<=20:
    i+=1
    if i%2==0:
        continue
    print ('Current value is ',i)
    
```

```python
# Use of pass 
for i in range(1,20,1):
    if i%2==0:
        print('inside pass value ',i)
        pass
    print(i)
```

# Function 

```python
def even_or_odd(num):
    if num%2==0:
        return "This is a Even number"
    else:
        return "this is a odd number"
```

```python
even_or_odd(13)
```

```python
def print_hello():
    print("Hello World")
```

```python
print_hello()
```

```python
# multiple opretation usimg function
def mul_and_sum(num1,num2):
    return f"Multiplication value is{num1*num2}",f"Sum value is {num1+num2}"
```

```python
mul_and_sum(12,24)
```

```python
# Default parameter
def userName(name='Guest'):
    return f"Username is {name}"
```

```python
userName('MAsum')
```

```python
# Postional Argument
def print_value(*values):
    for value in values:
        print(f"Value is {value}")
```

```python
print_value(1,2,3,46,7,89,4,'masum','mohua')
```

```python
# Keyword Argument
def print_item(**info):
    for key,value in info.items():
        print(f"{key}: {value}")
```

```python
print_item(name="Masum",age=24,city="Dhaka")
```
