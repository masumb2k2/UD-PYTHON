# Part two: Random Numbers

```python
np.eye(4) # 4x4 identity matrix
```

```python
np.random.rand(3,4) # random values in a 3x4 array between 0 and 1
```

```python
np.random.rand() # single random value between 0 and 1
```

```python
np.random.randn(4,5) # random values from standard normal distribution in a 4x5 array
```

```python
np.random.randint(10,21,size=(3,4)) # random integers between 10 and 20 in a 3x4 array
```

```python
rarr=np.random.randint(10,21,size=[5,2,3])
```

```python
rarr.shape # shape of the array
```

```python
rarr_2d=rarr.reshape(6,5)
```

```python
rarr.shape # shape of the array
rarr_2d.shape # shape of the reshaped array
```

```python
# usseful methods
rarr_2d.max() # maximum value in the array
```

```python
rarr_2d.min() # minimum value in the array
```

```python
rarr_1d=rarr_2d.reshape(30)
```

```python
rarr_1d.argmax()
```

```python
rarr_1d
```
