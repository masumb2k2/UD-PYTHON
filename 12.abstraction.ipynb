# abstracion

```python
from abc import ABC,abstractmethod
# abstract class
class Vehivle(ABC):
    def drive(self):
        print("The vehicle is used for driving")
    
    @abstractmethod
    def start_engine(self):
        pass
    
class Car(Vehivle):
    def start_engine(self):
        print("Car Engine Started")

def operate_vehivle(vechile):
    vechile.start_engine()

car=Car()

operate_vehivle(car)
```
