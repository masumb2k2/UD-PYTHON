# File operation- read & Write 

```python
# read a whole file 
with open('example.txt','r') as file:
    content=file.read()
    print(content)
```

```python
# read line by line 

with open('example.txt','r') as file:
    for line in file:
        print(line.strip()) # Sstrip remove the newline character
```

```python
# writting a file (overwrite)
with open('example.txt','w') as file:
    file.write('Hello Masum\n')
    file .write('This is a new line.\n')
```

```python
# write a file wihtout overwriting
with open('example.txt','a') as file:
    file.write('append operation\n')  
```

```python
# writting list of lines to a file 
lines=['First line\n','Second line\n','Third line\n']
with open('example.txt','a') as file:
    file.writelines(lines)
```

```python
# binary file operations
data=b'\x00\x01\x02\x03\x04\x05'
with open('example.bin','wb') as file:
    file.write(data)
```

```python
# read binary file
with open('example.bin','rb') as file:
    content=file.read()
    print(content)
```

```python
# read the content from source file and write to destination file
with open('example.txt','r') as src_file:
    content=src_file.read()
with open('destination.txt','w') as dest_file:
    dest_file.write(content)
```

```python
# writting & then reading a file
with open('example.txt','w+') as file:
    file.write('This is a test line.\n')
    # Move the cursor back to the beginning of the file
    file.seek(0)
    content=file.read()
    print(content)
```
