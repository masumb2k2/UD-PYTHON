```python
# Series
# A one-dimensional labeled array capable of holding any data type
data=[1,2,3,4,5]
series=pd.Series(data)
print(type(series))
series
```

```python
# create series from dictionary
data={'a':10,'b':20,'c':30}
series=pd.Series(data)
series
```

```python
data=[10,20,30]
index=['a','b','c']
pd.Series(data,index=index)
```

```python
# DataFrame
# A two-dimensional labeled data structure with columns of potentially different types
# create DataFrame from dictionary
data={
    'Name': ['Masum','mohua','Billah'],
    'age':[24,30,45],
    'city': ['dhaka','rajshahi','mirpur']
}
df=pd.DataFrame(data)
print(df)
print(type(df))
```

```python
# crate a datafram from list of dictionaries
data=[
    {'Name': 'Masum', 'age':24, 'city':'dhaka'},
    {'Name': 'mohua', 'age':30, 'city':'rajshahi'},
    {'Name': 'Billah', 'age':45, 'city':'mirpur'}
]
df=pd.DataFrame(data)
print(df)
print(type(df))
```
